```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read_csv - lê o arquivo CSV
#skiprows=8 - Ignora as primeiras 8 linhas do arquivo durante a leitura.
#delimiter=';' - Define o caractere que separa os campos em cada linha do arquivo.
#encoding='latin1'- indica que o arquivo está codificado em latin1.(outras codificações comuns incluem 'utf-8', 'utf-16'.)

df = pd.read_csv('INMET_MS_ITAQUIRAI_2020.CSV', delimiter=';', skiprows=8, encoding='latin1')
```

```python
#A função dtypes é uma ferramenta fundamental para entender os tipos de dados que compõem cada coluna de um conjunto de dados.
df.dtypes
```

```python
# Por padrão o head() mostra as 5 primeiras linhas do df.
df.head()
```

```python
#Por padrão o tail() mostra as 5 últimas linhas do df.
df.tail()
```

```python
#Reduzindo o df selecionando apenas as colunas que seram usadas
df = df[['Data','Hora UTC','PRECIPITAÇÃO TOTAL, HORÁRIO (mm)', 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)','TEMPERATURA DO PONTO DE ORVALHO (°C)','UMIDADE RELATIVA DO AR, HORARIA (%)', 'RADIACAO GLOBAL (Kj/m²)', 'VENTO, DIREÇÃO HORARIA (gr) (° (gr))' ,'VENTO, VELOCIDADE HORARIA (m/s)']]
```

```python
# Usando a função replace para substituir ',' por '.' 
df = df.replace(',','.',regex=True)
```

```python
#Verificando a alteração
#Métodos: São funções associadas a um objeto que realizam alguma ação sobre ele. Eles são chamados usando parênteses, como df.head().
df.head(15)
```

```python
#função isnull() detectar se há valores ausentes ou nulos em um df. Esses valores podem ser representados por NaN (Not a Number) para dados numéricos, None para objetos Python, ou NaT para datas.
#função sum() soma 

df.isnull().sum()
```

```python
#Transforma as colunas usando o FOR exceto as colunas 'Data' e 'Hora UTC' que devem manter em object.
#A função drop elimina as colunas data e hora UTC

coluna_objeto = df.select_dtypes(include=['object']).columns
coluna_objeto = coluna_objeto.drop(['Data', 'Hora UTC'])
for coluna in coluna_objeto:
    df[coluna] = pd.to_numeric(df[coluna], errors='coerce')

print(df.dtypes)
```

```python
# preenche todo os campos nulos ou vazio com 0
df = df.fillna(0)
```

```python
#Verificando se fez a alteração 
df.isnull().sum()
```

```python
#Normalizando coluna
df['UMIDADE RELATIVA DO AR, HORARIA (%)'] = (df['UMIDADE RELATIVA DO AR, HORARIA (%)'] - df['UMIDADE RELATIVA DO AR, HORARIA (%)'].min()) / (df['UMIDADE RELATIVA DO AR, HORARIA (%)'].max() - df['UMIDADE RELATIVA DO AR, HORARIA (%)'].min())
df
```

```python
#convertendo a data para data time
df['Data'] = pd.to_datetime(df['Data'], format='%Y/%m/%d')
```

```python
#Alterando o formato da data 
df['Data'] = df['Data'].dt.strftime('%d/%m/%Y')
```

```python
# Convertendo para datetime, assumindo que a data está no formato dd/mm/yyyy
df['Data'] = pd.to_datetime(df['Data'], format='%d/%m/%Y')

# Exibindo o DataFrame com a formatação desejada
print(df.to_string(formatters={'Data': lambda x: x.strftime('%d/%m/%Y')}))
```

```python
#Verificando tipo de dados 
df.dtypes
```

```python
#convertendo a hora
df['Hora UTC'] = pd.to_datetime(df['Hora UTC'], format='%H%M UTC', errors='coerce').dt.strftime('%H:%M')
```

```python
#Criando a coluna Hora BRT com fuso horário de Brasília
df['Hora BRT'] = pd.to_datetime(df['Hora UTC'], format='%H:%M').dt.tz_localize('UTC').dt.tz_convert('America/Sao_Paulo').dt.strftime('%H:%M')
```

```python
df.set_index('Hora BRT', inplace=True)

df[['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)', 'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)',
    'UMIDADE RELATIVA DO AR, HORARIA (%)']].plot(subplots=True)

plt.suptitle('Séries Temporais das Variáveis')
plt.show()
```

```python
df['Mês'] = pd.DatetimeIndex(df['Data']).month
meses_por_extenso = ['Janeiro', 'Fevereiro', 'Março', 'Abril', 'Maio', 'Junho', 'Julho', 'Agosto', 'Setembro', 'Outubro', 'Novembro', 'Dezembro']

df['Mês'] = df['Mês'].map(lambda x: meses_por_extenso[x-1])

# # Calculando a temperatura média Mensal
# media_mensal = df.groupby('Mês')['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].mean()
```

```python
df.head()
```

```python
# Dia mais quente e mais frio de cada mês
df_max_temp_dia = df.groupby(df['Data'])['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].max().reset_index()
df_min_temp_dia = df.groupby(df['Data'])['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].min().reset_index()
```

```python
# Criando o gráfico de linhas
plt.figure(figsize=(14, 6))
plt.plot(df['Mês'], df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], label='Temperatura')
plt.plot(df['Mês'], df['UMIDADE RELATIVA DO AR, HORARIA (%)'], label='Umidade')
plt.xlabel('Mês')
plt.ylabel('Valor')
plt.title('Variação de Temperatura, Umidade e Precipitação em 01/01/2020')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Agrupar por estação e calcular a precipitação média
df_grouped = df.groupby('Mês')['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'].mean()

plt.figure(figsize=(12, 6))
plt.bar(df['Mês'], df['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'], color='b')
plt.xlabel('Meses')
plt.ylabel('Precipitação')
plt.title('Média da precipitação mensal do ano 2020')
plt.tight_layout()
plt.show()
```

```python
# Temperatura mensal do ano 2020
df_grouped = df.groupby('Mês')['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'].max()

plt.figure(figsize=(12, 6))
plt.bar(df['Mês'], df['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)'], color='r')
plt.xlabel('Meses')
plt.ylabel('Temperatura')
plt.title('Soma da temperatura mensal do ano 2020')
plt.tight_layout()
plt.show()
```
